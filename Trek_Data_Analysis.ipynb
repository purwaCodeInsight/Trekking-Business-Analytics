```python
#change datatypes
df["booking_date"] = pd.to_datetime(df["booking_date"])
```

```python
df.info()
```

```python
#create columns for year and month
df["year"] = df["booking_date"].dt.year
df["month"] = df["booking_date"].dt.month_name()
```

```python
# age group column
bins = [17, 24, 34, 100]
labels = ["Young Adult", "Adult", "Senior"]

df["age_group"] = pd.cut(
    df["customer_age"],
    bins=bins,
    labels=labels
)
```

```python
df.head()
```

```python
#cancellation rate
cancellation_rate = (
    df["cancellation_status"].str.contains("Cancelled").sum()
    / len(df)
) * 100
print(cancellation_rate)
```

```python
#acrtual_revenue
refund_map = {
    "Completed": 1,
    "Partial Cancellation": 0.5,
    "Full Refund": 0
}

df["actual_revenue"] = (
    df["total_amount"] *
    df["cancellation_status"].map(refund_map)
).fillna(0)
```

```python
df.head()
```

```python
df["actual_revenue"].mean()
```

```python
#Group-wise Analysis
df.groupby("trek_location")["actual_revenue"].sum().sort_values(ascending=False)
```

```python
#Most prefered trek
df["trek_location"].value_counts()
```

```python
#Cancellation by Season
df.groupby("season")["number_of_people_cancelled"].sum()
```
